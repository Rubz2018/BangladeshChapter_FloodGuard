```python
import json
import os

# Get the current directory
current_path = os.getcwd()

# Construct the file path
file_path = os.path.join(current_path, 'report.json')

# Open the JSON file
with open(file_path) as file:
    # Load the JSON data
    data = json.load(file)



# Access the data
print(json.dumps(data, indent=2))
```

```python
print(type(data))
```

```python
collaborators = []
commits = []
lines_added = []


# Create lists to store the data
collaborators = []
commits = []
lines_added = []
lines_removed = []
active_days = []

# Iterate over the authors dictionary
for collaborator, info in data['authors'].items():
    collaborators.append(collaborator)
    commits.append(info['commits'])
    lines_added.append(info['lines_added'])
    lines_removed.append(info['lines_removed'])
    active_days.append(len(info['active_days']))

# Create a data frame
df = pd.DataFrame({
    'Collaborator': collaborators,
    'Commits': commits,
    'Lines Added': lines_added,
    'Lines Removed': lines_removed,
    'Active Days': active_days
})

# Print the data frame
print(df)
# Save the DataFrame to an Excel file

df.to_excel('./collaborator_data.xlsx', index=False)
```
