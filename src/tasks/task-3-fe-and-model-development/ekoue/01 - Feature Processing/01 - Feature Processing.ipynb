# Feature Engineering

The goal of this notebook is to create and test the code that will be used in the processing stage of Sagemaker Pipeline. 

```python
# Todo: add this code in sagemaker processing code

import pandas as pd
import numpy as np
import os
```

```python
# Todo: add this code in sagemaker processing code

BASE_DIR= "../../../"

DATA_FOLDER= os.path.join(BASE_DIR, "data/")
DAILY_WEATHER_FILE= os.path.join(DATA_FOLDER, "weather/ekoue/combined/divisions daily weather.csv")

CLEANED_DAILY_WEATHER_FILE= os.path.join(DATA_FOLDER, "weather/ekoue/cleaned/cleaned divisions daily weather.csv")

TRAIN_DAILY_WEATHER_FILE= os.path.join(DATA_FOLDER, "weather/ekoue/cleaned/sagemaker divisions daily weather -- train.csv")
VAL_DAILY_WEATHER_FILE= os.path.join(DATA_FOLDER, "weather/ekoue/cleaned/sagemaker divisions daily weather -- val.csv")
TEST_DAILY_WEATHER_FILE= os.path.join(DATA_FOLDER, "weather/ekoue/cleaned/sagemaker divisions daily weather -- test.csv")


DAILY_WEATHER_FILE, CLEANED_DAILY_WEATHER_FILE, TRAIN_DAILY_WEATHER_FILE, VAL_DAILY_WEATHER_FILE, TEST_DAILY_WEATHER_FILE
```

```python
# Todo: add this code in sagemaker processing code

weather_df = pd.read_csv(DAILY_WEATHER_FILE)
weather_df.head(3).T
```

```python
weather_df.columns.to_list()
```

```python
# Todo: add this code in sagemaker processing code

# Columns to keep in the dataset (removing those that may induce target leakage )
wanted_columns = [
     'cloudcover',
     #'conditions',
     'Date',
     #'DayType',
#      'description',
     'dew',
     'Division',
     'feelslike',
#      'feelslikemax',
#      'feelslikemin',
     'humidity',
     'moonphase',
#      'precip',
#      'precipprob',
#      'precipcover',
#      'preciptype',
     'sealevelpressure',
#      'severerisk',
     'Season',
     'snow',
#      'snowdepth',
     'solarenergy',
     'solarradiation',
#      'stations',
     'temp',
     'sunset',
     'sunrise',
     'tempmax',
     'tempmin',
     'uvindex',
#      'visibility',
     'winddir',
#      'windgust',
     'windspeed',
     'Flood',
     'Sunrise Hour',
     'Sunrise Minute',
     'Sunset Hour',
     'Sunset Minute'
]

weather_df = weather_df[wanted_columns]
weather_df.head(3).T
```

```python
weather_df.shape
```

```python
# Todo: add this code in sagemaker processing code

# Remove samples with null season (those are outside of our range of study anyway, which is Jan 2013 to Oct 2023)
weather_df.dropna(subset=["Season"], inplace=True)
weather_df.shape
```

```python
weather_df.head(3).T
```

```python
weather_df.info()
```

```python
# Todo: add this code in sagemaker processing code

# Convert date columns to datetime type
weather_df["Date"] = pd.to_datetime(weather_df["Date"])
weather_df["sunset"] = pd.to_datetime(weather_df["sunset"])
weather_df["sunrise"] = pd.to_datetime(weather_df["sunrise"])

weather_df.info()
```

```python
weather_df.head(3).T
```

```python
# # Todo: add this code in sagemaker processing code

# weather_df["Date"] = weather_df["Date"].dt.date
# weather_df["sunset"] = weather_df["sunset"].dt.time
# weather_df["sunrise"] = weather_df["sunrise"].dt.time

# weather_df.head(3).T
```

```python
weather_df.info()
```

```python
# Todo: add this code in sagemaker processing code

print(weather_df.shape)
weather_df.dropna(inplace=True)
weather_df.shape
```

```python
# Todo: add this code in sagemaker processing code

# dummify categorical data
weather_df = pd.get_dummies(weather_df, columns=["Division","Season"], prefix=["Division","Season"])
weather_df.head(3).T
```

```python
# Todo: add this code in sagemaker processing code

# Engineering new columns
weather_df["DayOfMonth"] = weather_df["Date"].dt.day
weather_df["Month"] = weather_df["Date"].dt.month
weather_df["Year"] = weather_df["Date"].dt.year

# get elapsed time between sunrise and sunset, in minutes
weather_df["sunrise_sunrise_mindiff"] = ( weather_df["sunset"] - weather_df["sunrise"]).dt.total_seconds()/60

weather_df.head(3).T
```

```python
# Todo: add this code in sagemaker processing code

# Drop irrelevant columns to the project 
columns_to_drop = [
    "snow", # has too many null values
    "Date", # no more needed, as features were already engineered based on it
    "sunset", # no more needed, as features were already engineered based on it
    "sunrise", # no more needed, as features were already engineered based on it
]
weather_df.drop(columns_to_drop,axis=1,inplace=True)
```

The following cells will put the label as the first column in the df

```python
# Todo: add this code in sagemaker processing code
y = weather_df.pop("Flood")
y.head()
```

```python
# Todo: add this code in sagemaker processing code

weather_df.insert(0, "Flood", y)
weather_df.head(3).T
```

```python
weather_df["Flood"].value_counts()
```

```python
# Todo: add this code in sagemaker processing code

# Replace True and False by 1 and 0
weather_df["Flood"] = weather_df["Flood"].apply(lambda x: 1 if x==True else 0)
weather_df["Flood"].value_counts()
```

```python
# Todo: add this code in sagemaker processing code

# shuffling dataframe content
weather_df = weather_df.sample(frac = 1, random_state=51294).reset_index(drop=True)
```

```python
weather_df.shape
```

```python
# Todo: add this code in sagemaker processing code

# Compute split lenghts
train_len = int(weather_df.shape[0] * .7)
val_len = int(weather_df.shape[0] * .15)
test_len = weather_df.shape[0] - (train_len + val_len)

train_len, val_len, test_len, train_len+val_len+test_len==weather_df.shape[0]
```

```python
# Todo: add this code in sagemaker processing code

# Split datasets in train, val, test
train_df = weather_df.loc[:train_len-1, :]
test_df = weather_df.loc[train_len: train_len+val_len-1, :]
val_df = weather_df.loc[train_len+val_len:, :]

train_df.shape, test_df.shape, val_df.shape, train_df.shape[0]+test_df.shape[0]+ val_df.shape[0] == weather_df.shape[0]
```
