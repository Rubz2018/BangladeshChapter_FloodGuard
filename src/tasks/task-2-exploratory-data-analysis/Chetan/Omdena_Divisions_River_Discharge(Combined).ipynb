```python
df.set_index(df.columns[0], inplace=True)
df.info()
```

```python
# Check for duplicates based on 'river_discharge' and 'Year'
duplicates = df[df.duplicated(subset=['river_discharge', 'Year'], keep=False)]

# Print the number of duplicates
print(f'Number of duplicates: {len(duplicates)}')
```

```python
df= df.drop_duplicates(subset=['river_discharge', 'Year'], keep='first')
df.shape
```

```python
df = df.sort_values(by=['Year', 'river_discharge'])
null_percentage=round(df.isnull().sum() / len(df) * 100, 2)
```

```python
null_percentage
```

```python
# Drop the columns with more than 80% null values

threshold = 80  # Set the threshold for null values

# Get the columns that have more than 80% null values
columns_to_drop = null_percentage[null_percentage > threshold].index

df.drop(columns=columns_to_drop, inplace=True)
```

```python
columns_to_drop
```

```python
df.shape
```

```python
round(df.isnull().sum() / len(df) * 100, 2)
```

```python
# Skewness of columns with null values (Numeriam Values):

skewness_values = df[[ 'visibility', 'windspeed_x', "sealevelpressure"]].skew()
skewness_values
```

```python
# Remove the NaN values and replace with Median:

columns_with_null = ['visibility', 'windspeed_x', 'sealevelpressure']

df[columns_with_null] = df[columns_with_null].fillna(df[columns_with_null].median())
```

```python
# Check if "precipitation_sum" and "rain_sum" are the same:

are_same = df['Year'].equals(df['Day'])
print(f'precipitation_sum and rain_sum are the same: {are_same}')
```

```python
# Mean "river_discharge" group by district and year ,  hue="year" - Excluding "rajshahi"
filtered_df = df[df['Year'] != 'Month']
average_river_discharge_name_year=filtered_df.groupby(["Year", "Month"])[["river_discharge"]].aggregate('mean')
average_river_discharge_name_year = average_river_discharge_name_year.reset_index()

plt.figure(figsize=(15, 8))
sns.set_palette("viridis")

colors = sns.color_palette("viridis", n_colors=len(average_river_discharge_name_year["Year"].unique()))

sns.barplot(x="Month", y="river_discharge", hue="Year", data=average_river_discharge_name_year, palette=colors)
plt.xlabel("Month", fontsize=16)
plt.ylabel("Mean river_discharge", fontsize=16)
plt.title("Mean River Discharge", fontsize=16)
#plt.xticks(rotation=45)
plt.xticks(fontsize=14)
plt.tight_layout()
plt.legend(title="Year",  bbox_to_anchor=(1.02, 1), loc="upper left")
plt.show()
```
