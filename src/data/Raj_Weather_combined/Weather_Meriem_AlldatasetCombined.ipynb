Merging all the collumns (weather and river discharge data) together and merging rows of all districts together

```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
cwd
```

```python
file_list = os.listdir(cwd+'/Weather_data')
```

```python
file_list
```

```python
for filename in file_list:
  df_district = pd.read_csv(f'Weather_data/{filename}',index_col='datetime',parse_dates=True)
  df_district['city'] =filename.split('_')[0]
  df_district.to_csv(f'/content/{filename}')
```

```python
df = pd.concat( map(pd.read_csv, ['Rajshahi_Weather_combined.csv',
 'Dhaka_Weather_combined.csv',
 'Sylhet_Weather_combined.csv',
 'Khulna_Weather_combined.csv',
 'Mymensigh_Weather_combined.csv',
 'Barisal_Weather_combined.csv',
 'Naryanganj_weather_combined.csv',
 'Comilla_Weather_combined.csv',
 'Chittagong_weather_combined.csv',
 'Bogra_Weather_combined.csv']), ignore_index=True)
```

```python
df.to_csv('Weather_allCities.csv', index=False)
```
